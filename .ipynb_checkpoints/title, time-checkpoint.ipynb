```python
import requests
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
from urllib import parse
import time
```

```python
head_url = "https://www.10000recipe.com/recipe/list.html?q="

menu_list = []
for i in cate_1:
    menu_list.append(head_url + parse.quote(i))
print(menu_list)
```

```python
result_01 = []
for i in menu_list:
    
    recipe_urls_list = []
    for j in range(1, 5): # 711

        page = urlopen(i + "&order=reco&page=" + str(j))

        soup = BeautifulSoup(page, "lxml")
        
        recipe_urls = []
        for x in range(40):

            recipe_urls.append(soup.select("div.common_sp_thumb a")[x].get("href"))
        
        recipe_urls_list.extend(recipe_urls)
        
    result.extend(recipe_urls_list)
            
    print(len(recipe_urls_list))
```
